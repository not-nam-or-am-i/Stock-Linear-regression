```python
import numpy as np
import matplotlib.pyplot as plt
```

# File này dùng để học cách sử dụng thư viện plt

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('simple plot')
ax.set_ylabel('amplitude')
ax.set_xlabel('time')
```
